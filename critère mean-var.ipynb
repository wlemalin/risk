# Calcul V_p :

```r
library(stringr)

#rendements
r_1 <- 5
r_2 <- 4

#écart-type
sd1 <- 3
sd2 <- 4

#coef. de corr.
ρ <- 0

#répartition PF (part actif 1)
θ <- 0.5

# Def Fonction V(μ,σ²)
V. <- function(r_P, var_P){
    (r_P) - 0.2*(var_P)
}
```

```r
var1 <-sd1**2
var2 <- sd2**2
r_P <- θ*r_1 + (1-θ)*r_2
var_P <- (θ**2)*var1 + ((1-θ)**2)*var2 + 2*ρ*θ*(1-θ)*sd1*sd2
sd_P <- sqrt(var_P)
```

```r
print("Stats du PF :")
print(str_c("Rentabilité espérée = ",r_P))
print(str_c("Standard deviation = ",sd_P))
print(str_c("Utility V(P) = ",V.(r_P,var_P)))
```

# Calcul θ optimal :

```r
Vpmax <- 0
for (i in 0:1000){
    θ <- i/1000
    r_P <- θ*r_1 + (1-θ)*r_2
    var_P <- (θ**2)*var1 + ((1-θ)**2)*var2 + 2*ρ*θ*(1-θ)*sd1*sd2
    
    if (V.(r_P,var_P)>= Vpmax){
        Vpmax <- V.(r_P,var_P)
        θ_max <- θ       
    }
}

print(str_c("V(P) maximum = ",Vpmax))
print(str_c("θ argmax V(P) = ",θ_max))
```
