# Calcul DS1

```r
library(stringr)
sf = 5 #Sig.Fig.
```

```r
##################################################
pL1 = c(0.6,0.35,0.05)            #Proba loterie F

pL2 = c(0.6,0.2,0.2)             #Proba loterie G
##################################################
```

```r
if(length(pL1)==length(pL2)){
    N <- length(pL1)
}else{
    N <- ("ERROR")
}
FL1 = GL2 = c(rep(0,N))
FL1[1] <- pL1[1]
GL2[1] <- pL2[1]
for (i in 2:N){
    FL1[i] <- pL1[i] + FL1[i-1]
    GL2[i] <- pL2[i] + GL2[i-1]
}
n <- 0
for (i in 1:N){
    if(round(FL1[i]-GL2[i],sf)==0){
        n=n+1
    }
}
print(str_c("Nb. réalisations possibles = ",N))
print(FL1)
print(GL2)
print(str_c("Nb. de valeures égales : ",n))
```

```r
FdomG <- 0
GdomF <- 0

for (i in 1:N){
    if(round(FL1[i]-GL2[i],sf)>0){
        GdomF=GdomF+1
    }
}
    if(GdomF==N-n){
        print("No but G DS1 F")
    }
for (i in 1:N){
    if(round(FL1[i]-GL2[i],sf)<0){
        FdomG=FdomG+1
    }
}
    if(FdomG==N-n){
        print("Yes F DS1 G")
    }

print(str_c("Score F vs G : ",FdomG,"/", GdomF))
```
